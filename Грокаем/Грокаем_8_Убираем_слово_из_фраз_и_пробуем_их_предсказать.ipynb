```python
# список списков в которых каждое слово - отдельный элемент через включая пробелы и тире
# типа: [['rff','we3'],['jkh','ooiu']]
tokens = list(map(lambda x:(x.split(" ")), raw_reviews))
```

```python
tokens
```

```python
wordcnt = Counter() # выдает словарь в котором ключ - это слово в переданном списке,
                    # а значение - это сколько раз данное слова появляется в списке
for sent in tokens:
  for word in sent:
    wordcnt[word] -= 1
# в данном случае передается словарь и подсчет ведется для всех слов ваще (почему-то в обратную сторону)
```

```python
wordcnt['insane']
```

**most_common([n]) - возвращает n наиболее часто встречающихся элементов, в порядке убывания встречаемости. Если n не указано, возвращаются все элементы.**

**>>> Counter('abracadabra').most_common(3)**

**[('a', 5), ('r', 2), ('b', 2)]**

```python
wordcnt.most_common()
```

```python
# короче здесь получается список слов от наиболее часто встречающихся к менее часто
vocab = list(set(map(lambda x:x[0], wordcnt.most_common())))
```

```python
vocab
```

```python
# создает словарь: ключ - наиболее встречающееся слово,
                # значение - порядковый номер
word2index = {}
for i,word in enumerate(vocab):
  word2index[word]=i
```

```python
word2index
```

```python
# здесь создается список списков, где каждый список - это комментарий, здесь каждое слово переведено в цифру
# т.е все комментарии состоят из цифр, каждое слово в любом комментарии имеет свою цифру
concatenated = list()
input_dataset = list()
for sent in tokens:
  sent_indices = list()
  for word in sent:
    try:
      sent_indices.append(word2index[word])
      concatenated.append(word2index[word])
    except:
      ""
  input_dataset.append(sent_indices)
concatenated = np.array(concatenated)
```

```python
concatenated # получили список кодов слов, чем меньше число, тем чаще встречается слово
```

```python
def similar(target='beautiful'):
  target_index = word2index[target] # здесь заложена цифра слова 'beautiful' = 2023

  scores = Counter()
  for word,index in word2index.items():
    raw_difference = weights_0_1[index] - (weights_0_1[target_index]) # т.е. ((вектор 1х100) - 2023)
    squared_difference = raw_difference * raw_difference # получим вектор 1х100
    scores[word] = -math.sqrt(sum(squared_difference)) # math.sqrt(sum(squared_difference)) - число
  return scores.most_common(10) # выводит 10 самых часто встречающихся
```
