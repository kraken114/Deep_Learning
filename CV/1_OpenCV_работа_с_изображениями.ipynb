# **Код для видео**

```python
cap = cv2.VideoCapture('/content/gdrive/MyDrive/Colab Notebooks/OPENCV/VID_20190502_120115.mp4')
```

**создаем цикл показа картинок одной за другой из кот складывается видео:**

```python
from google.colab.patches import cv2_imshow
```

```python
while True:
  success,img = cap.read() #success - true/falls(читается/не читается видео)
  cv2_imshow(img) # показывает изображение
  if cv2.waitKey(5000) & ord('q') == 0xFF: # cv2.waitKey - время отображения картинок, 'q' - для выхода из картинки
    break
```
